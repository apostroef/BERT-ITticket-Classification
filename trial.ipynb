```python
from Library import *

df = pd.read_csv("translated_data.csv")
tiket_df = pd.DataFrame(df)

from scipy.stats import chi2_contingency

for i in range(1, 9):
    col = f"tag_{i}"
    if col in tiket_df.columns:  # check column exists
        contingency = pd.crosstab(tiket_df["priority"], tiket_df[col])
        chi2, p, dof, expected = chi2_contingency(contingency)
        print(f"{col}: Chi-square = {chi2:.2f}, p-value = {p:.4f}")


def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    r, k = confusion_matrix.shape
    return np.sqrt((chi2 / n) / (min(k - 1, r - 1)))


print("Cramer's V:", cramers_v(tiket_df["priority"], tiket_df["body"]))
```
